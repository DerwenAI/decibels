```python
df = pd.read_csv("Soundcheck2.csv")
```

```python
postal = sorted(df.postal_code.unique())       
ypos=np.arange(len(postal))
decibels = df.groupby("postal_code").avg_decibels.mean()
plt.bar(ypos, decibels, align='center')
plt.xticks(ypos, postal, rotation=80, fontsize=3)
plt.ylabel('Average Decibels')
plt.title('Average Decibels vs Postal Code')
plt.show()

```

```python
postal = sorted(df.postal_code.unique())       
ypos=np.arange(len(postal))
decibels = df.groupby("postal_code").max_decibels.mean()
plt.bar(ypos, decibels, align='center')
plt.xticks(ypos, postal, rotation=80, fontsize=3)
plt.ylabel('Max Decibels')
plt.title('Average Max Decibels vs Postal Code')
plt.show()
```

```python

locations = sorted((df.primary_category_name.unique()))       
ypos=np.arange(len(locations))
decibels = df.groupby("primary_category_name").max_decibels.mean()
plt.bar(ypos, decibels, align='center')
plt.xticks(ypos, locations, rotation=80, fontsize=4)
plt.ylabel('Max Decibels')
plt.title('Average Max Decibels vs Type of Venue')
plt.show()
```

```python
sat_avg=df[df.local_day_of_week=="sat"]["avg_decibels"].mean()
sun_avg=df[df.local_day_of_week=="sun"]["avg_decibels"].mean()
mon_avg=df[df.local_day_of_week=="mon"]["avg_decibels"].mean()
tue_avg=df[df.local_day_of_week=="tue"]["avg_decibels"].mean()
wed_avg=df[df.local_day_of_week=="wed"]["avg_decibels"].mean()
thu_avg=df[df.local_day_of_week=="thu"]["avg_decibels"].mean()
fri_avg=df[df.local_day_of_week=="fri"]["avg_decibels"].mean()
plt.bar("Sat", sat_avg) 
plt.bar("Sun", sun_avg) 
plt.bar("Mon", mon_avg) 
plt.bar("Tue", tue_avg) 
plt.bar("Wed", wed_avg)
plt.bar("Thu", thu_avg)
plt.bar("Fri", fri_avg) 
plt.title(" Average Decibels vs Day of Week")
plt.ylabel("Decibels")
```

```python
df['datetime'] = pd.to_datetime(df['local_time'])
y = pd.DatetimeIndex(df.datetime)
x=y.hour+(y.minute/60)
df['Decimal_time'] = x

Time = sorted(df.Decimal_time.unique())       
ypos=np.arange(len(Time))
decibels = df.groupby("Decimal_time").avg_decibels.mean()
plt.bar(ypos, decibels, align='center')
plt.xticks(ypos, Time, rotation=80, fontsize=0)
plt.ylabel('Average Decibels')
plt.title('Average Decibels vs Time')
plt.show()
```
